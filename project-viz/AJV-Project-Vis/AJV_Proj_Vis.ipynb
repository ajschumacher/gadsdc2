```python
import pandas
import numpy
from matplotlib import pylab
%matplotlib inline
from scipy import stats
```

```python
oppdata1 = pandas.read_csv('AJV_Prioritized_CSV_v2_COPY.csv')
```

```python
# NAICS is short for North American Industry Classification System, which is an indexed scheme of commercial goods and services.
# I'm using this to identify where most of my prioritized opportunities fall within this categorization.

pylab.plot(oppdata1.NAICS,oppdata1.Priority,'.')
pylab.ylim(0,5.5)
pylab.xlabel('NAICS Code')
pylab.ylabel('Priority (1-5 Scale)')
pylab.title('Plot 1: Priority vs NAICS Classification')
```

```python
pylab.plot(oppdata1.Contract_Val,oppdata1.Priority,'.')
pylab.xlim(0,500000)
pylab.ylim(0,5.5)
pylab.xlabel('Contract Value (x $1000)')
pylab.ylabel('Priority (1-5 Scale)')
pylab.title('Plot 2: Priority vs Contract Value (Shortened)')
```

```python
#Translate Contract Types into indexable integer values
Contract_Type_Index = set(oppdata1.Contract_Type)
Contract_Type_Dictionary = zip(range(len(Contract_Type_Index)),Contract_Type_Index)
Contract_Type_Key = numpy.zeros(len(oppdata1.Contract_Type))
for item in range(len(oppdata1.Contract_Type)):
    for i in range(len(Contract_Type_Index)):
        if (oppdata1.Contract_Type[item] == Contract_Type_Dictionary[i][1]):
            Contract_Type_Key[item] = Contract_Type_Dictionary[i][0]
readable = pandas.DataFrame(Contract_Type_Dictionary)
print readable[1]
```

```python
Contract_Type_Freq_Table = stats.itemfreq(Contract_Type_Key)
pylab.plot(Contract_Type_Freq_Table.T[1],'o')
pylab.xlabel('Contract Type Category (See Legend)')
pylab.ylabel('Frequency of Occurrences')
pylab.title('Plot 3: Frequency Chart of Contract Types')
```
